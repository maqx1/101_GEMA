## 1. 三角函数

#### 1. 角的：“角度表示” 与 “弧度表示”

高中时，我们学习过 $$\sin 30^\circ=\frac{1}{2}$$$$\sin 45^\circ=\frac{1}{\sqrt{2}}$$$$\sin 60^\circ=\frac{\sqrt{3}}{2}$$

1. 这里的 $45^\circ$ 是角的**角度表示**。
2. 为了更方便地处理角度，人们用角在**单位圆**上扫过的**弧长**表示角的大小，称为角的**弧度表示**。
$$360^\circ=2\pi$$
$$1^\circ=\frac{2\pi}{360}=\frac{\pi}{180}$$$$\pi=180^\circ$$
4. 微积分中，如无特殊说明，我们 **一般采用弧度** 表示，因此上述关系写为
$$\sin \frac{\pi}{6}=\frac{1}{2}$$$$\sin \frac{\pi}{4}=\frac{1}{\sqrt{2}}$$$$\sin \frac{\pi}{3}=\frac{\sqrt{3}}{2}$$

#### 2. 三角函数对称性 

1. 奇偶性 $$\sin x=-\sin (-x),\ \ \ \cos x=\cos(-x)$$
2. 对称性 $$\sin x=\sin (\pi-x),\ \ \ \cos x=-\cos(\pi-x)$$
3. 周期性 $$\sin x=\sin(2\pi+x),\ \ \ \cos x=\cos(2\pi+x)$$
4. 对偶性 $$\sin x=\cos\left(\frac{\pi}{2}-x\right),\ \ \ \cos x=\sin\left(\frac{\pi}{2}-x\right)$$
5. 勾股定理 $$\sin^2x+\cos^2x=1$$

#### 3. 和角公式

$$\cos(\alpha+\beta)=\cos\alpha\cos\beta-\sin\alpha\sin\beta$$
$$\sin(\alpha+\beta)=\sin\alpha\cos\beta+\cos\alpha\sin\beta$$

#### 4. 欧拉公式 

对于任何弧度角$\alpha$，我们定义复数$$f(\alpha)=\cos\alpha+\sqrt{-1}\sin\alpha$$

我们按和角公式展开 $$f(\alpha+\beta)=\cos(\alpha+\beta)+\sqrt{-1}\sin(\alpha+\beta)$$
$$=\cos\alpha\cos\alpha-\sin\alpha\sin\beta+\sqrt{-1}\sin\alpha\cos\beta+\sqrt{-1}\cos\alpha\sin\beta$$
$$=(\cos\alpha+\sqrt{-1}\sin\alpha)(\cos\beta+\sqrt{-1}\sin\beta)=f(\alpha)\cdot f(\beta)$$

角度的加法，变成了对应复数的乘法。这很像指数运算 $$x^{a+b}=x^a\cdot x^b$$

欧拉定义了指数函数：$$e^z=1+z+\frac{z^2}{2!}+\frac{z^3}{3!}+\cdots$$
在此定义下，他证明了以下公式， $$e^{i\alpha}=\cos\alpha+i\sin\alpha$$
在这个意义下，$f$真的成为了指数运算。上述公式称为**三角函数的欧拉公式**

#### 5. 和差化积、积化和差 

1. 和差化积$$\cos\alpha+\cos\beta=\cos\left(\frac{\alpha+\beta}{2}+\frac{\alpha-\beta}{2}\right)+\cos\left(\frac{\alpha+\beta}{2}-\frac{\alpha-\beta}{2}\right)=2\cos\left(\frac{\alpha+\beta}{2}\right)\cos\left(\frac{\alpha-\beta}{2}\right)$$
$$\sin\alpha+\sin\beta=\sin\left(\frac{\alpha+\beta}{2}+\frac{\alpha-\beta}{2}\right)+\sin\left(\frac{\alpha+\beta}{2}-\frac{\alpha-\beta}{2}\right)=2\sin\left(\frac{\alpha+\beta}{2}\right)\cos\left(\frac{\alpha-\beta}{2}\right)$$

2. 积化和差$$\cos\alpha\cos\beta=\frac{\cos(\alpha-\beta)+\cos(\alpha+\beta)}{2}$$
$$\sin\alpha\cos\beta=\frac{\sin(\alpha+\beta)+\sin(\alpha-\beta)}{2}$$


#### 6. 等差数列的三角函数求和

$$\sin x+\sin 2x+\cdots+\sin nx=?$$

1. 利用欧拉公式$$\sin kx=\frac{e^{ikx}-e^{-ikx}}{2i}$$
$$\sum_{k=1}^n\sin kx=\sum_{k=0}^n\sin kx=\mathrm{Im}\left(\sum_{k=0}^ne^{ikx}\right)=\mathrm{Im}\left(\frac{1-e^{i(n+1)x}}{1-e^{ix}}\right)$$
$$=\mathrm{Im}\left(\frac{e^{-i\frac{n+1}{2}x}-e^{i\frac{n+1}{2}x}}{e^{-i\frac{x}{2}}-e^{i\frac{x}{2}}}\right)\left(\frac{e^{i\frac{n+1}{2}x}}{e^{i\frac{x}{2}}}\right)=\mathrm{Im}\left(\frac{\sin-\frac{n+1}{2}x}{\sin-\frac{x}{2}}\right)e^{\frac{n}{2}x}$$$$=\left(\frac{\sin\frac{n+1}{2}x}{\sin\frac{x}{2}}\right)\sin\frac{n}{2}x=\frac{-\cos (n+\frac{1}{2})x+\cos\frac{1}{2}x}{2\sin\frac{x}{2}}$$
2. 用和差化积裂项$$\sin kx\sin\frac{1}{2}x=\frac{1}{2}\left(-\cos (k+\frac{1}{2})x+\cos(k-\frac{1}{2})x\right)$$
两边从$1$到$n$求和即得$$S_n\cdot\sin\frac{1}{2}x=\frac{1}{2}\left(-\cos(n+\frac{1}{2})x+\cos\frac{1}{2}x\right)$$

```python
%matplotlib qt
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")

[x,t] = np.meshgrid(np.array(range(25))/24.0,np.arange(0,500,0.5)/500*10*np.pi-2*np.pi)

p = 2*np.exp(-t/(8*np.pi))
u = np.cos(3.2*t)**2
y = (x**2-x)**2*np.sin(p)
r = u*(x*np.sin(p)+y*np.cos(p))

surf = ax.plot_surface(r*np.cos(t), r*np.sin(t), u*(x*np.cos(p)-y*np.sin(p)),
                       rstride=1, cstride=1, cmap=cm.Wistia, #viridis/plasma/inferno
                       linewidth=0,antialiased=True)

plt.show()
```

$$x\in [0,1],\ t\in[-2\pi,8\pi],\ p=2 e^{-\frac{t}{8\pi}},\ u=\cos(3.2t)^2$$
$$y=(x^2-x)^2\cdot\sin(p),\ r=u\cdot(x\sin p+y\cos p)$$
$$(X,Y,Z)=(r\cos t,\ r\sin t,\ u\cdot(x\cos p-y\sin p))$$
