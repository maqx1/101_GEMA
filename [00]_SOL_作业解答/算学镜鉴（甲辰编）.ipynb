## **例案甲编** 商贾退课事
距期末之试仅五日。有学子谒师而问曰：“弟子今岁甚忙，与友创业于南海之滨，近日公司方立，无暇温课备考。恐试不及格，有碍保研，可否退课？”

## **例案乙编** 赌试失备考

岁次壬寅，《高数》授业首五章并十一章。试毕，有生诣师曰："吾忖度当试首六章，必不涉十一章，故专攻第六章。岂意今反试十一章，愿师垂怜，赐予及格，可乎？"

<font color=blue>按</font>：此生未尝趋讲堂，自同门处得旧岁试题。岂料今试迥异于昔，殊为憾也。

## **例案丙编** 旷课应试案

某生，上学年《高等算学》课业获评甲等。及至下学岁，竟日赴衙署实习（晨昃至酉），诸课皆自修。每观例题，自谓已解，然少习演，终致试绩在常品之下。

<font color=blue>按</font>：夫为学者，当以**笃实**为本。此子虽颖悟能自课，然废师徒面授之益，绝同窗切磋之机，譬犹习射而弗引弓，诵谱而绝抚弦也。昔人云："熟能生巧"，今观其例，益信然矣。且夫实习虽要，岂宜尽夺诵习之时？《学记》有言："时过然后学，则勤苦而难成"，此之谓也。为学之道，贵在**循序而渐进**，专一而致精。若首鼠两端，终致两失，可不慎欤？

## **例案丁编** 云窗堕算科

试毕高数，惟求速死（掩面）（伏地不起）。

自问： 此半岁何为？
自答： 祸起一寐——高数早课，酣眠至隅中（爽哉！），遂寻《b站大学》补之（倍速观课，更爽哉！）。

冬月寒甚，遂决意： 闭户从「b站夫子」游。
b站诲余： 「孺子，行路当如是。」
余拊掌： 「易耳！」
及试，题云： 「请舞《国标》一段。」
余茫然： 「……行路之法，竟为何耶？」

后询同窗，方知堂上早习此舞！
b站误我！

然祸不单行——时令乖舛，尤甚于此。
普化败绩，遂昼夜扑于实验录、反复啃噬经籍，高数功课，草草塞责。偶忆则启b站，敷衍数题（呜呼！此时竟忘高数乃四分重课，普化仅三）。

今忏悔： 此试使吾本孱弱之算学，雪上加霜。来岁必改！

<font color=blue>按</font>：然此子下学期仅至半期，期中试毕，遂绝迹于讲堂。

嗟乎！此子之怠，可谓甚矣。夫学业者，如逆水行舟，不进则退。始以高数甲等之资，而终至旷课废学，犹持千金之璧而投诸渊也。昔闻"靡不有初，鲜克有终"，今观此子，信然。且夫期中既试，便即辍学，是犹耕者见苗而不溉，行者半途而止息。子曰："譬如为山，未成一篑，止，吾止也。"此之谓欤？

## **例案戊编** 苟完害学案

期中既过，课业更制，令习书末之题。然题尾既有解答，自校即可知得失，毋需呈师批阅。纵强令上传，不过检视完否而已。

且「书末习题」之制，吾常草录于杂记，或径书于卷侧，字迹狼藉，不便摄传。由是虽成其业，终未上传。

然此「完成」二字，实不堪细究——答对者但钩之，错者不诘其由，难者朱笔誊答案而已，未尝深究正法。洎乎考场，睹多熟题，竟唯忆初答之谬，而正道茫然！

<font color=blue>按</font>： 此子之失，正合《学记》"独学而无友，则孤陋而寡闻"之戒。自校不严，如沐猴而冠；抄录答案，犹乞儿拾金——虽持宝而不知所用也！

吾今悟其失者有二：

其一，究习题之义理，本当归于课业之常经。昔者吾以"题毕即功成"，此心实谬，今当深省。

其二，高等算学者，确需竭智尽力，课余尤当自课。向者吾持"分毫不肯多费"之念，岂非自误？今亦痛悔。

<font color=blue>按</font>： 此子之悔，暗合古人"学然后知不足"之理。若能持此二悟，则前失可转为后得，正所谓"不贵于无过，而贵于能改过"（王阳明语）。

## **例案己编** 学算觉悟篇

余尝惑焉：彼时百思不解者，旬日后再观，竟豁然开朗。若罗尔定理一章，初学时但觉名目纷繁，逻辑缠杂；及至期末温习，则脉络粲然，不复解昔日何以困顿至此。期中试卷亦然，期末重做时，殊不解曩日何以谬误频频。

究其本因：

- 一者，废预习之良习。虽中学时尝持之，然入大学后，谬以为"光阴充裕，何须预读"，殊不知：无预习则听讲如坠云雾。积压既多，复为它课作业、会社事务所困。

- 二者，半期前心志弛懈。盖大学之道异于中学：疑而不问，则惑终不解。但求形似，实则蒙昧

<font color=blue>按</font>： 当初觉得困难，实因功夫未到；后来能够理解，正是学力渐深。若不预习，犹如无桨行于急流；敷衍应付，好比盲人持烛夜行。大学之道，贵在自强不息，岂能因循怠惰，自误前程？大学者，存疑不诘则永惑，苟且趋功则难精

《预学箴》：凡课业，必先预习。非求精熟，然需明其纲目。盖大学讲席，异于庠序：

- 罕有演题之隙（不似中学多习题时）
- 但传精要之旨（师所授皆浓缩）
- 况早课昏晨（高数线代皆在"朝昃"）
- 困倦交攻（睡意与难理相搏）

<font color=blue>按</font>："不预则如夜行无烛，虽遇明珠而不识"

## **例案庚编** 算学得失录

所学之难：
定理相济为用，限、积、级数诸要义尤艰。

所长：
习算甚勤，于积分之道颇有所得。暇时尝遍索异题于网，虽多奇技淫巧，然实能活思广智。

所短：
经籍未精，不能触类旁通。盖因但知其然，未究其所以然，后当穷理尽性，务求数理之真。

总论：
然观此学期，进学之心甚炽，既拓思路，复得良法，获益实多。

<font color=blue>按</font>：此子能明得失，可谓"知耻近乎勇"（《中庸》）。其以"活思"为乐，暗合程颐"深思而自得"之旨。若能持此热忱而补原理之缺，则"虽愚必明，虽柔必强"矣。

## **例案辛编** 期末复课录

温课要旨：
专攻积分与级数二章。然收敛半径之端点判据，温习时草草掠过，致试卷屡谬于此。

级数得失：
- 泰勒、幂级数温之尤勤
- 求和函数时，初循导积常法而困蹇
- 临试终刻方悟次数参差之法，然时已不逮
- 傅里叶级数误判其易，遂疏于习

进学状态：
季初涉新学，畏其难而勤勉；季末生倦意，然考测示警后即振。

吾之弊，常在：心志易弛（如弓久张则疲），诸课并进时左支右绌。虽有过失，然自课之志未堕，每察其失必速矫之。

<font color=blue>按</font>："不积跬步，无以至千里"，今虽得积分之妙，然级数未通，犹未足矜也。来岁四课并修，当效董子"下帷讲诵"，庶几无愧于"学然后知不足"之训。



## **例案壬编** 课徒记异

有生来学而不听讲，自陈"未达"，遂踞末席，日纂《考研捷录》为业。后与之语，见其叩问皆失鹄，懵于理本，惟以记题之术演算耳。

<font color=blue>按</font>： 今之塾间多此"画荻生"：面壁而趋时，抱椟而遗珠。昔郑板桥讥"抄书吏"，康南海讽"应举囚"，正为此辈画像。彼辈岂知算题若无理贯，犹航沧溟而无斗杓，虽日演千题，终不免舛驳支离之患。为师者当以"启牖"为急，先破其机械之心，而后可语格致之道。

## **例案癸编** 习算自讼

积分数术，吾所习实未精熟。若积分之表、有理式之属，本当强记以速算，然未能尽诵。致应试之时，运笔迟滞，又因遗忘关窍，临推无术，遂致谬误迭出。

<font color=blue>按</font>：今之习算者，多务速成而忽根本，犹筑台不基而求其高。昔梅文鼎《历算全书》曰："数理虽繁，必由本达末；算法虽捷，必因熟生巧。"盖积算一道，非徒恃记诵，亦贵明其理而娴其术。若但囫囵强背，遇新题则束手；倘能通变，纵未熟记，亦可推演得之。学者当以穷理为本，辅以缮性，庶几事半功倍矣。
