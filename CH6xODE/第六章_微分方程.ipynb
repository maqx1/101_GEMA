$$ \textrm{\hspace{1cm} \ 方程\hspace{.3cm}}\frac{\mathrm{d}^2x}{\mathrm{d}t^2}=-\frac{k}{m}x+c\frac{\mathrm{d}x}{\mathrm{d}t}$$

通解根据阻尼强度，分三种情况

  -a. 欠阻尼： $c^2<4mk$, 
       $$x(t)=e^{-\frac{c}{2m}t}(A\cos\omega t+B\sin\omega t),$$$$\textrm{其中常数：} \omega=\sqrt{\frac{k}{m}-\left(\frac{c}{2m}\right)^2}$$
  -b. 临界阻尼：$c^2=4mk$，$$x(t)=(A+Bt)e^{-\frac{c}{2m}t}$$
  -c. 过阻尼：$c^2>4mk$, 
$$x(t)=Ae^{\lambda_1t}+Be^{\lambda_2t},$$$$\textrm{其中常数：}\lambda_1=-\frac{c}{2m}+\sqrt{\left(\frac{c}{2m}\right)^2-\frac{k}{m}},\ \lambda_2=-\frac{c}{2m}-\sqrt{\left(\frac{c}{2m}\right)^2-\frac{k}{m}}$$

如果固定$m=k=1$，初始位移为$1$, 初始速度为$-1$, 在欠阻尼、临界取阻、过阻尼，分别取阻尼系数 $c=1,2,3$ 则关于方程的解，可以画出以下图像。
1. $c=1$, $x(t)=e^{-t/2}(A\cos \frac{\sqrt{3}}{2}t+B\sin \frac{\sqrt{3}}{2}t)$, $A=1,B=-\frac{\sqrt{3}}{3}$
2. $c=2$, $x(t)=(A+Bt)e^{-t}$, $A=1,B=0$
3. $c=3$, $x(t)=Ae^{\frac{-3+\sqrt{5}}{2}t}+Be^{\frac{-3-\sqrt{5}}{2}t}$, $A=\frac{5+\sqrt{5}}{10},B=\frac{5-\sqrt{5}}{10}$

```python
import numpy as np
import matplotlib.pyplot as plt

## 中文显示设置，根据不同系统，可咨询DeepSeek辅助寻找相关配置
font_name = 'STSong'  # 可替换为你想测试的字体名称
plt.rcParams['font.family'] = font_name

## 以下考虑 k=1,m=1 的情况

def f1(x): ## 欠阻尼, c=1
    return np.exp(-x/2)*(np.cos(np.sqrt(3)*x/2)-np.sin(np.sqrt(3)*x/2)/np.sqrt(3))
def f2(x): ## 临界阻尼, c=2
    return np.exp(-x)
def f3(x): ## 过阻尼, c=3
    return np.exp((-3+np.sqrt(5))/2*x)*(5+np.sqrt(5))/10+np.exp((-3-np.sqrt(5))/2*x)*(5-np.sqrt(5))/10
def f4(x): ## 基准线 x=0
    return 0*x
# 在(0,10)之间等距采样1000个数据点绘图
x = np.linspace(0,10,1000)

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
y4 = f4(x)

plt.figure(figsize=(8,6))
plt.plot(x, y1, color="blue", label="欠阻尼 c=1") 
plt.plot(x, y2, color="red", label="临界阻尼 c=2")
plt.plot(x, y3, color="green", label="过阻尼 c=3")
plt.plot(x, y4, color="black", label="参考基准x=0")

plt.title("根据不同阻尼系数c，方程关于x(0)=1,x'(0)=-1的定解")
plt.xlabel("时间 t")
plt.ylabel("位移 x")
plt.legend()
plt.show()
```

## **1. 微分方程的基本概念**

1. 微分方程：联系`若干自变量`、`若干未知函数`、`未知函数的导数或微分`的等式，称为<font color=red>微分方程</font>。
   1. **[A0101]** 自变量$x$, 未知函数$y$, `导数`的等式  $$3x(y')^2-4y^2y'=x$$
   2. **[A0102]** 自变量$x$, 未知函数$y$, `微分`的等式  $$(e^{-x}-x^3y)\mathrm{d}x-2x\mathrm{d}y=0$$
   3. **[p360, 例5]** 自变量$t$, 未知函数$S,I,R$，表示易感染人群(S)，已感染人群(I)，康复人群(R)，满足微分方程组
      $$\left\{\begin{aligned}
\frac{\mathrm{d}S}{\mathrm{d}t}&=-aSI,\\
\frac{\mathrm{d}R}{\mathrm{d}t}&=bI,\\
\frac{\mathrm{d}I}{\mathrm{d}t}&=aSI-bI.\\
\end{aligned}\right.$$
   4. **[课外阅读]** 自变量$x,y,t$, 未知函数 $I$的（偏）微分方程。
      $$I_x \frac{\mathrm{d}x}{\mathrm{d}t}+I_y \frac{\mathrm{d}y}{\mathrm{d}t}+ I_t=0$$
      手机运行游戏时，使用<font color=blue>光流方程</font>插帧，(以提高画面帧数，减少运动残影）。其中$x,y$是像素坐标，$I(x,y,t)$是$t$时刻像素点亮度（其中记号$I_x$
      $$I_x=\frac{\partial I}{\partial x}=\lim_{h\to 0}\frac{I(x+h,y,t)-I(x,y,t)}{h}$$是一个关于$x,y,t$的函数，称为$I$关于$x$的偏导数，详见以下视频第四节：https://www.bilibili.com/video/BV1zsr5Y7Ecy/?share_source=copy_web

2. 常微分方程：注意到上节A,B,C都只有一个自变量。我们称只含有`一个自变量`的微分方程，为<font color=red>常微分方程</font>。不然，称之为偏微分方程，如D。


3. 常微分方程的阶：方程中出现的关于未知函数的最高阶导数，称为常微分方程的<font color=red>阶</font>。

  4. 常微分方程的解：若函数$y=\phi(x)$代入微分方程$F(x,y,...,y^{(n)})$=0成立，则称函数$\phi$为<font color=red>微分方程的解</font>。
     1. 例如 $\phi(x)=e^x$ 是方程 $y''=y$ 的解。注意，这样的方程只告诉你未知函数是$y$, 但没有指明自变量是$t$ 或者$x$，则需要用向读者<font color=red>简要说明</font>自变元记号。
     2. 注意到 $2e^x，e^{-x}$都是方程的解，因此方程的解不唯一。解的表达式若含有参数，且<font color=red>参数个数等于方程阶数$n$</font>，称为<font color=red>通解</font>，如$y''=y$通解为$Ae^x+Be^{-x}$.
     3. 不含未知参数的解称为<font color=red>特解</font>，例如$2e^x$是特解。含未知参数，但未知参数个数小于阶数，例如 $Ae^x+2e^{-x}$, 我们只能称为$y''=y$的一族解。
     4. 从通解确定特解一般需要$n$个条件，称为<font color=red>定解条件</font>。从通解求特解的问题称为<font color=red>定解问题</font>。给出在某个点的$n$个导数,称为<font color=red>初值条件</font>或柯西条件,也有指定不同点的值或导数值，称为边值条件。一般，只要定解条件总个数等于方程总阶数，就可以定解（详见[603]中阻尼问题）。一般来说，常微分方程的初值条件解存在且唯一，但边值条件解未必总是存在（想象扔一个橡皮鸭子）。

5. Python 编程实践：解逻辑斯蒂(logistic)方程 **[p318]** $$\frac{\mathrm{d}P}{\mathrm{d}t}=rP-\beta P^2$$
   这是荷兰人口学家Verhulst提出，名称logistic来自希腊语logistikos，意为“合理的”, 描述<font color=blue>受限增长</font>过程。

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(x,r,b):
    return r*x-b*x**2

r = 1.0 # 增长率
b = 0.1 # 资源限制系数

dt = 0.1 # 迭代步长
steps = 50 # 迭代次数
t = np.linspace(0,dt*steps,steps+1) # 时间轴采点
```

```python
def plotlogistic(P0): # 绘制初始人口为P0的解 
    P = np.empty(steps+1)
    P[0]=P0
    
    for i in range(steps): # 欧拉折线法迭代
        P[i+1] = P[i] + logistic(P[i],r,b)*dt
        
    plt.plot(t,P,color="blue") # 绘图
```

```python
for j in np.arange(0,20,1):
    plotlogistic(j)
```

```python
P=np.empty(steps+1)
plt.plot(t,P*0+r/b,color="red") # 人口收敛到极限值 r/b, 为什么？

#plotlogistic(-1)
plt.show()
```

```python
#%matplotlib qt
%matplotlib inline
"""
以上是 jupyter notebook 专属命令，控制输出图像为可拖动弹窗，或静态行嵌入notebook
"""

import matplotlib.pyplot as plt
import numpy as np

def lorenz(xyz, *, s=10, r=28, b=2.667):
    """
    对象 xyz : 3元数组，代表点的空间坐标，注意这里x,y,z是天气系统的状态量（如压强、温度等）
    对象 s,r,b : 洛伦茨方程中的参数
    返回对象 xyz_dot : 3元数组，方程决定该点的空间坐标的变化率
    """
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])


dt = 0.01
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))  
"""
对象 xyzs 是 numsteps+1 行 3 列数组，这里+1是为了设定初始值
"""

xyzs[0] = (0., 1., 1.05)  # Set initial values
# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point

"""
经过 dt = 0.01时间间隔的 num_steps = 10000 步迭代，
"""

for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt

ax = plt.figure().add_subplot(projection='3d')
ax.plot(*xyzs.T, lw=0.5)
"""
* 是 Python 中的 解包操作符（unpacking operator）。它的作用是将一个可迭代对象（如列表、元组、数组等）解包为独立的参数
"""

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```
