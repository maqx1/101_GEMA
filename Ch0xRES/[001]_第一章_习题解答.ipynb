### [603] 能力 ：建立直觉（<font color=blue>学有余力的同学</font>）

1. 常微分方程的学习不仅仅是分类和计算，还需要理解方程背后的<font color=blue>物理意义、几何意义</font>以及解的性质（如稳定性、周期性等）。单纯依赖分类和计算可能会忽略对问题的深入理解。例如，你能看出以下方程描述什么物理现象、解中参数的意义、能从函数图像分析解的物理含义吗?

    1. 简谐振动，圆周运动 $$ \textrm{\hspace{1cm} \ 方程\hspace{.3cm}}\frac{\mathrm{d}^2x}{\mathrm{d}t^2}=-x, \textrm{\hspace{1cm} \ 通解\hspace{.3cm}}x(t)=C_1\sin t+C_2\cos t,$$
    2. 代入物理常数的简谐振动，如弹性系数为 $k$, 滑块质量为 $m$ 的轻质弹簧$$\textrm{\hspace{1cm} \ 方程\hspace{.3cm}}\frac{\mathrm{d}^2x}{\mathrm{d}t^2}=-\frac{k}{m}x, \textrm{\hspace{1cm} \ 通解\hspace{.3cm}}x(t)=C_1\sin\sqrt{\frac{k}{m}}t+C_2\cos \sqrt{\frac{k}{m}}t,$$
    3. 受阻力的弹簧，假设阻力与速度成正比（如水中的弹簧）
$$ \textrm{\hspace{1cm} \ 方程\hspace{.3cm}}\frac{\mathrm{d}^2x}{\mathrm{d}t^2}=-\frac{k}{m}x+c\frac{\mathrm{d}x}{\mathrm{d}t}$$

通解根据阻尼强度，分三种情况

  -a. 欠阻尼： $c^2<4mk$, 
       $$x(t)=e^{-\frac{c}{2m}t}(A\cos\omega t+B\sin\omega t),$$$$\textrm{其中常数：} \omega=\sqrt{\frac{k}{m}-\left(\frac{c}{2m}\right)^2}$$
  -b. 临界阻尼：$c^2=4mk$，$$x(t)=(A+Bt)e^{-\frac{c}{2m}t}$$
  -c. 过阻尼：$c^2>4mk$, 
$$x(t)=Ae^{\lambda_1t}+Be^{\lambda_2t},$$$$\textrm{其中常数：}\lambda_1=-\frac{c}{2m}+\sqrt{\left(\frac{c}{2m}\right)^2-\frac{k}{m}},\ \lambda_2=-\frac{c}{2m}-\sqrt{\left(\frac{c}{2m}\right)^2-\frac{k}{m}}$$

如果固定$m=k=1$，初始位移为$1$, 初始速度为$-1$, 在欠阻尼、临界取阻、过阻尼，分别取阻尼系数 $c=1,2,3$ 则关于方程的解，可以画出以下图像。
1. $c=1$, $x(t)=e^{-t/2}(A\cos \frac{\sqrt{3}}{2}t+B\sin \frac{\sqrt{3}}{2}t)$, $A=1,B=-\frac{\sqrt{3}}{3}$
2. $c=2$, $x(t)=(A+Bt)e^{-t}$, $A=1,B=0$
3. $c=3$, $x(t)=Ae^{\frac{-3+\sqrt{5}}{2}t}+Be^{\frac{-3-\sqrt{5}}{2}t}$, $A=\frac{5+\sqrt{5}}{10},B=\frac{5-\sqrt{5}}{10}$

```python
import numpy as np
import matplotlib.pyplot as plt

## 中文显示设置，根据不同系统，可咨询DeepSeek辅助寻找相关配置
font_name = 'STSong'  # 可替换为你想测试的字体名称
plt.rcParams['font.family'] = font_name

## 以下考虑 k=1,m=1 的情况

def f1(x): ## 欠阻尼, c=1
    return np.exp(-x/2)*(np.cos(np.sqrt(3)*x/2)-np.sin(np.sqrt(3)*x/2)/np.sqrt(3))
def f2(x): ## 临界阻尼, c=2
    return np.exp(-x)
def f3(x): ## 过阻尼, c=3
    return np.exp((-3+np.sqrt(5))/2*x)*(5+np.sqrt(5))/10+np.exp((-3-np.sqrt(5))/2*x)*(5-np.sqrt(5))/10
def f4(x): ## 基准线 x=0
    return 0*x
# 在(0,10)之间等距采样1000个数据点绘图
x = np.linspace(0,10,1000)

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
y4 = f4(x)

plt.figure(figsize=(8,6))
plt.plot(x, y1, color="blue", label="欠阻尼 c=1") 
plt.plot(x, y2, color="red", label="临界阻尼 c=2")
plt.plot(x, y3, color="green", label="过阻尼 c=3")
plt.plot(x, y4, color="black", label="参考基准x=0")

plt.title("根据不同阻尼系数c，方程关于x(0)=1,x'(0)=-1的定解")
plt.xlabel("时间 t")
plt.ylabel("位移 x")
plt.legend()
plt.show()
```
